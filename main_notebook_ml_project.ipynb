# Abstract

_Large email inboxes could be automated using flexible, pre-trained Machine Learning models. This requires NLP techniques that perform feature extraction and dimensionality reduction on text, in conjunction with other available data provided by the customer. This research focuses on available supervised and unsupervised  techniques to approach this automation, and explores BERT, Google's first unsupervised, deeply bidirectional system for pre-training NLP.

# Introduction

## Introduction and research topic

**Research topic**

*Process automation of large email inboxes using Machine Learning*

ML6 is a Belgian Machine Learning company who is developing a product that automates parts of big email inboxes, e.g. typical info@ email inboxes. The company has split the problem into two parts. Firstly, a classifier is to be trained to correctly classify emails into labels and automate the processing of these emails. Secondly, relevant categories of emails are to be found.

There exists a sizeable dataset of complaints about consumer financial products and services that can be used to train and develop a Machine Learning model. The goal of this project is to make it easier to process these complaints. A description of the various fields in this dataset can be found [here](https://cfpb.github.io/api/ccdb/fields.html).

**Introduction**

Email inboxes, e.g. typical info@ email addresses, are occasionally still maintained by people. These systems become more difficult and eventually infeasible to manage as email quantities increase. Parsers are used to filter emails and forward them to the correct endpoint, e.g. emails regarding payments are sent to the financial department and emails regarding paid leave are sent to HR. Parser generally use rule-base systems, which make them very inflexible and susceptible to errors ([source](https://mailparser.io/blog/what-is-an-email-parser)). More flexible solutions are thus needed to address this issue. 

Precision could be increased using Machine Learning models. It is trained on a freely available collection of complaints about consumer financial products and services ([source](https://www.consumerfinance.gov/data-research/consumer-complaints)), which serve as the dataset. Firstly, using supervised learning, a classifier is developed to classify emails. This is split into two phases. The first phase focuses solely on learning via *Consumer Complaint Narratives*, which are the complaint texts written by the customer. The second phase of supervised training adds other columns from the dataset to the model, in an attempt to increase its accuracy in predicting the right label. Secondly, through unsupervised learning  categories of emails are classified. Thirdly and finally, Google's BERT (Bidirectional Encoder Representations from Transformers) ([source](https://arxiv.org/pdf/1810.04805.pdf)) is used to help label emails.

**SVMs** are a type of supervised machine learning models. They are fast and dependable and offer a relatively simple approach to the NLP classification problem at hand. These SVMs are very well suited for text-categorization ([Joachims, 1998](http://www.cs.cornell.edu/people/tj/publications/joachims_98a.pdf)). A linear kernel is recommended ([Stecanella, 2017](https://monkeylearn.com/blog/introduction-to-support-vector-machines-svm/)), because text is often linearly separable. This kernel is excellent when there are a significant amount of features, as mapping the data to a higher dimensional space does not really improve the performance ([Kowalczyk, 2014](https://www.svm-tutorial.com/2014/10/svm-linear-kernel-good-text-classification/)).

A **Random Forest** is a collection of many decision trees and serves as a classifier to solve multi classification tasks ([Pranckevicius et al, 2017](https://www.bjmc.lu.lv/fileadmin/user_upload/lu_portal/projekti/bjmc/Contents/5_2_05_Pranckevicius.pdf)). Rather than just simply averaging the prediction of trees, which could be called a forest, the model uses two key concepts that gives it the name random ([source](https://towardsdatascience.com/an-implementation-and-explanation-of-the-random-forest-in-python-77bf308a9b76)):
* Random sampling of training data points when building trees
* Random subsets of features considered when splitting nodes

Every decision tree has the same nodes, but has different leaves, because they are using different data. Th model then merges the decisions of multiple decision trees in order to find an answer, which represents the average of all these decision trees ([source](https://medium.com/capital-one-tech/random-forest-algorithm-for-machine-learning-c4b2c8cc9feb)).

<img src="images/random_forest_1.png" alt="random_forest_1" width="300"/>

Every node has a Gini Impurity, which is the probability that a randomly chosen sample in a node will be incorrectly classified according to the distribution of samples in the node. When this value decreases, it signifies that a node is becoming more pure. This means that there is a lower probability that a complaint randomly selected from the dataset would be misclassified.

It should be noted that Random Forests are susceptible to outliers and missing data ([source](https://medium.com/capital-one-tech/random-forest-algorithm-for-machine-learning-c4b2c8cc9feb)). Therefore, the data must be cleaned thoroughly before-hand.

### Data Cleaning functions

```python
import re
import string

# Remove urls
def clean_url(complaint):
    # to do: more regex url garbage matching
    complaint = re.sub('https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+', '', complaint)
    complaint = re.sub('https? ?: ?// ?(?:[-\w.]|(?:%[\da-fA-F]{2}))+', '', complaint)
    return complaint

# Remove punctuation from complaint
def clean_punctuation(complaint):
    complaint = re.sub('[%s]' % re.escape(string.punctuation), '', complaint)
    return complaint

# Remove non-sensical characters from complaint
def clean_nonsense(complaint):
    complaint = re.sub('[''""...]', '', complaint)
    complaint = re.sub('\n', '', complaint)
    return complaint

# Remove censored words from complaint
def clean_censored(complaint):
    complaint = re.sub('[XXXX]', '', complaint)
    return complaint

# Turn everything into lowercase
def clean_lowercase(complaint):
    complaint = complaint.lower()
    return complaint

# Remove numbers from complaint
def clean_numbers(complaint):
    complaint = re.sub("\d+", '', complaint)
    return complaint
```

### Lemmatization

[Nltk](https://www.nltk.org/book/)'s *WordNetLemmatizer* was used to lemmatize the complaint texts. Recent research indicates that stemming and lemmatization outperform  baseline algorithms ([Balakrishnan et al, 2014](https://www.researchgate.net/publication/271293133_Stemming_and_Lemmatization_A_Comparison_of_Retrieval_Performances)). Stemming could be considered as an alternative method to lemmatization, however this would increase recall, due to generalization, and harm precision ([source](https://nlp.stanford.edu/IR-book/html/htmledition/stemming-and-lemmatization-1.html), [source](https://nlp.stanford.edu/IR-book/html/htmledition/evaluation-of-ranked-retrieval-results-1.html#p:map)).  Lemmatization produces better precision compared to stemming ([Balakrishnan et al, 2014](https://www.researchgate.net/publication/271293133_Stemming_and_Lemmatization_A_Comparison_of_Retrieval_Performances)). However, the differences are virtually insignificant for the English language. Other European languages show a more significant percent change in mean average precision from stemming and lemmatization, with diacritic removal gains up to 23% ([Salton, 1989](http://www.iro.umontreal.ca/~nie/IFT6255/Introduction.pdf)) ([Hull, 1996](http://citeseerx.ist.psu.edu/viewdoc/download?doi=10.1.1.16.8310&rep=rep1&type=pdf)) ([Hollink et al, 2004](https://link.springer.com/article/10.1023/B:INRT.0000009439.19151.4c)). Additional cleaning steps could include dealing with typos, but this was out of scope for these experiments.

The cleaned complaints are now tokenized. This operation removes stop words, such as *and*, *a(n)* and *the*, and allows for three hyper-parameter to be tuned: <code>ngram_range</code> (bigrams, trigrams, up to "n"grams), <code>min_df</code> (only keep words that appear in a minimum of x documents) and <code>max_df</code> (only keep words that appear in a maximum of x% of documents). The result of this operation is a Bag of Words, which is a Document-Term Matrix. TF-IDF is then applied, which is used as a scoring measure to reflect how relevant a term is in a given *Consumer complaint narrative*.

# Sprint 2: unsupervised learning

This sprint will explore unsupervised learning algorithms, as well as a new method of dimensionality reduction, called Doc2Vec. This is an NLP tool for representing documents as a vector and is a generalisation of the Word2Vec method ([source](https://towardsdatascience.com/multi-class-text-classification-with-doc2vec-logistic-regression-9da9947b43f4)). Section [3.1](#Exploratory-Data-Analysis-(EDA)), **Exploratory Data Analysis (EDA)**, gave a deep insight into the data and its various columns. The data was then **cleaned** in section [3.2](#Data-Cleaning) with **NLP techniques**, such as removal of punctiation and unnecessary characters, lemmatization and tokenization. Additional **Data Preprocessing** was performed in section [3.3](#Data-Preprocessing-(for-other-columns-in-the-dataset)) to thoroughly analyse the various other columns in the dataset. 

A Bag of Words (BoW) was constructed and TF-IDF was applied to this Document-Term Matrix. In sprint 2, this BoW will be replaced with word embeddings, which are a more advanced technique. Technically speaking, word embeddings can be seen as a method of dimensionality reduction, because they map words with similar meaning to similar points in a high dimensional space. The word embedding technique we have chosen to explore in this sprint is **Doc2Vec**. As previously mentioned, this is a generalisation of the Word2Vec method and allows us to represent documents as a vector. 

In 2013, [Mikolov et al](https://arxiv.org/pdf/1301.3781.pdf) described **Word2Vec** and proposed two new architectures to compute continuous vector representations of words from very large data sets, namely **Continuous Bag-of-Words** and **Continuous Skip-gram**. They found that these techniques can help answer subtle semantic relationships between words for high dimensional word vectors. These two-layer neural network based word vectors allow for much better results than previous state of the art techniques and can be trained on a virtual unlimited size of vocabulary, with a significant increase in speed as well.

While Word2Vec computes a feature vector for every word in the corpus, **Doc2Vec computes a feature vector for every document in the vector**. This allows us to determine which documents (complaints) are similar. Word2Vec works on the inituation that the word representation should be good enough to predict the surrounding words, while Doc2Vec's underlying intuition is that the document representation should be good enough to predict the words in the document ([source](https://www.quora.com/What-is-doc2vec)).

Take the following example ([source](https://techscouter.blogspot.com/2018/08/doc2vec-document-vectorization-and.html)):

* John has taken many leaves the year.
* Leaves are falling of the tree.

In these two sentences, the word "leaves" has a different meaning based on the sentence in which it is used. This meaning can only be captured when we are taking the context of the complete phrase into account or if we would like to measure the similarity of the phrases and cluster them under one name.

According to [Le and Mikolov](https://arxiv.org/pdf/1405.4053.pdf), the document vector is an **unsupervised algorithm** that learns fixed-length feature representations from variable-length pieces of text, such as sentences, paragraphs and documents. Their algorithm represents each document by a dense vector which is trained to predict words in the document. Similarly to Word2Vec, Doc2Vec can be implemented in two ways: **PV-DM (Paragraph Vector Distributed Memory) or PV-DBOW (Paragraph Vector Distributed Bag of Words)**.

Socher et al. have also focused on distributed representations of phrases and sentences. Their techniques, however, are supervised and thus require more labeled data to work well. Paragraph Vector, in contrast, is mostly unsupervised and thus can **work well with less labeled data** ([source, p. 7](https://arxiv.org/pdf/1405.4053.pdf)).

### Data preprocessing

It is worth noting that **Mikolov et al performed nearly no data preprocessing**, such as lemmatization, stemming and removing stop words, on the dataset in their paper. There is **no real consensus** on whether or not it is good practice to remove stop words or to stem words for Word2Vec and Doc2Vec. It mostly depends on what one is trying to achieve. For short texts or sentences, such as titles or tweets, it may be a good idea to remove stop words because their frequency could be very high compared to key words. In our case, the complaints are generally slightly longer than a tweet. More experiments could be done to see if removing stop words, numbers and/or lemmatizating words would offer improvements.

### Doc2Vec

We will now use these CSV files as input for our Doc2Vec algorithm. [Gensim](https://radimrehurek.com/gensim/models/doc2vec.html)'s implementation of Doc2Vec expects our consumer complaint narratives to be of class TaggedDocuments. 

**We refer to [this](Doc2Vec.ipynb) notebook (Doc2Vec.ipynb) for the Doc2Vec model training. Running this notebook will train the DBOW model. The code below serves as an example of how Gensim can be used to train the Doc2Vec DBOW model on one of our imbalanced datasets and one of our balanced datasets. The consumer complaint narratives are the same for the **cp, cps and cpsi** variants of our dataset. Thus, training the DBOW model only needs to occur twice. We then save this model and use it in our clustering algorithms.

```python
import pandas as pd
import numpy as np
from gensim.models.doc2vec import Doc2Vec, TaggedDocument

# imbalanced
df_imbalanced = pd.read_csv("corpus_sprint2_imbalanced_cp.csv", encoding="utf-8")
# balanced
df_balanced = pd.read_csv("corpus_sprint2_balanced_cp.csv", encoding="utf-8")

# There appears to be 1 narrative that is null. Drop it.
df_imbalanced = df_imbalanced.dropna()
df_balanced = df_balanced.dropna()
```

Next, we have to tokenize our consumer complaint narratives.

```python
import nltk
from nltk.corpus import stopwords

def tokenize_text(text):
    tokens = []
    for sent in nltk.sent_tokenize(text):
        for word in nltk.word_tokenize(sent):
            if len(word) < 2:
                continue
            tokens.append(word.lower())
    return tokens
```

The input for a Doc2Vec model should be a list of TaggedDocument(['list','of','words'], [TAG]). We could give every document a unique ID (such as a sequential serial number) as a document tag, or a shared string tag representing something else about it, or both at the same time. We have chosen to pass the Product as a tag.

```python
tagged_imbalanced = [TaggedDocument(words=tokenize_text(complaint), tags=[i]) for i, complaint in enumerate(df_imbalanced["Consumer complaint narrative"])]

tagged_balanced = [TaggedDocument(words=tokenize_text(complaint), tags=[i]) for i, complaint in enumerate(df_balanced["Consumer complaint narrative"])]
```

[Lau and Baldwin](https://arxiv.org/pdf/1607.05368.pdf) provide recommendations on hyper-parameter settings for general-purpose applications.

The hyperparameters that were tuned by Lau and Baldwin are:
* vector size: the dimension of word vectors;
* window size: left/right context window size;
* min count: minimum fequency threshold for word types;
* sub-sampling: threshold to downsample high frequency words;
* negative sample: number of negative word samples;
* epoch: number of training epochs. More iterations take more time and eventually reach a point of diminishing returns.

They did not tune the initial and minimum learning rates (α and α<sub>min</sub>, respectively), and used the the following values for all experiments: α = .025 and α<sub>min</sub> = .0001. The learning rate decreases linearly per epoch from the initial rate to the minimum rate ([source](https://arxiv.org/pdf/1607.05368.pdf)).

Creating the model and setting its parameters:
* We set <code>dm = 0</code>, which means we are using DBOW. The feature vectors are obtained by training a neural network on the task of predicting a probability distribution of words in a paragraph given a randomly-sampled word from the paragraph ([source](https://arxiv.org/pdf/1607.05368.pdf)).
* We set <code>window = 2</code>, which is the maximum distance between the current and predicted word within a sentence.
* We set <code>min_count = 2</code>, which means we ignore all words with a total frequency lower than this number.
* We set <code>vector_size = 50</code>, which means we choose the dimensionality of the generated feature vectors.
* We set <code>alpha = 0.025</code> as the initial learning rate, which will linearly drop to <code>min_alpha = 0.0001</code> as training progresses.
* We train it for 20 epochs, <code>epochs = 20</code>.
* <code>workers = 4</code> allows us to use 4 worker threads.

We can use multiprocessing to speed up learning.

```python
import multiprocessing
cores = multiprocessing.cpu_count()
```

Progress bar:

```python
from tqdm import tqdm
tqdm.pandas(desc="progress-bar")
```

#### Creating the model & building the vocabulary

```python
# create the model (one for each type of dataset)
model_imbalanced = Doc2Vec(dm=0, vector_size=50, window=2, min_count=2, alpha= 0.025, min_alpha=0.0001, epochs=20, workers=cores)
model_balanced = Doc2Vec(dm=0, vector_size=50, window=2, min_count=2, alpha= 0.025, min_alpha=0.0001, epochs=20, workers=cores)

# build the vocabulary (one for each type of dataset)
model_imbalanced.build_vocab([x for x in tqdm(tagged_imbalanced)])
model_balanced.build_vocab([x for x in tqdm(tagged_balanced)])
```

#### Training

We have chosen to train for 20 epochs, because a value of 10-20 is common in published work ([source](https://radimrehurek.com/gensim/auto_examples/tutorials/run_doc2vec_lee.html)) when dealing with tens-of-thousands to millions of documents. 

```python
model_imbalanced.train(tagged_imbalanced, total_examples=model_imbalanced.corpus_count, epochs=model_imbalanced.epochs)
```

#### Save the model

## KMeans

### Introduction

K-means clustering is an iterative method of classifying vectors into k different classes. Each of these vectors is assigned to the nearest _mean_ of a class, splitting the dimensional data into __Voronoi cells__. The algorithm scales the data so that each dimension has approximately the same magnitude. 

The clustering of the data works as follows:
1. The datapoints are randomly assigned to classes. One way of doing this is choosing k random datapoints as means and assigning the remaining datapoints to the nearest means. In this case, step 1.A. can be skipped in the first iteration.
2. Repeat the following steps until no changes accur anymore or the maximum amount of iterations is reached
    1. Calculate the mean of the new classes
    2. Assign the datapoints to their nearest mean
    3. If there is a group with too little elements, eliminate this group and assign them to the remaining means
    
<img src="images/kmeans.PNG" alt="kmeans" width="500"/>

Once this algorithm has reached a result, the classes are not necessarily optimal. The magnitude of these classes could be very divergent. One way of tackling this problem is repeating the algorithm with different random starting points for the means, and keeping the one that had the best distribution. 

The question then remains what the value of k should be. One way of determining the optimal k value, is to iterate over different values of k. When starting with k = 0 and iteratively increasing it's value, the variation in the different classes will go down. Then a point where the variation does not reduce as drastically as it dit before is determined. The k at this turning point is then chosen as the optimal k value. This method is called the _elbow method_. Another way of doing this is calculating the silhouette score of the classes and choosing the k value that results in the highest score. 

In this project the decision was made to use the _elbow_ method to search for interesting k values. These values were then fed to functions that calculated the silhouette score and the maximum was determined.  

One remark that should be made is that the algorithm does not guarantee to find the optimum. This is because it doesn't review all possible combinations, as it is an NP-hard algorithm that uses heuristics to find local optimums.

[source](https://en.wikipedia.org/wiki/K-means_clustering), [source](https://www.youtube.com/watch?v=4b5d3muPQmA)

## DBSCAN

### Introduction

__DBSCAN__ or __Density-based spatial clustering of applications with noise__ is, as the name suggests, a density-based clustering algorithm. Spatial points that are packed close together are assigned to the same class. The algorithm also takes the outliers into account that lie in low-density areas. 

DBSCAN uses four different types of _points_ to classify the data. These are:
- __Core points__ are points that lie within the specified radius of at least a certain amount (also to be specified) of other points
- A point a is a __directly reachable point__ from core point b, lies within the specified radius of point b
- Point a is __reachable__ from core point b if there is a path of directly reachable core points between them
- All other points are __outliers__ or **noise points**.   

To form a cluster, points must be mutually density-connected (directly reachable in both directions) to at least the specified minimum amount of other datapoints. Other datapoints that are directly reachable from at least one point in the cluster, are also taken as part of that cluster.

The algorithm works as followed:
1. An arbitrary starting point is chosen and its $\epsilon$-neighborhood is determined. The $\epsilon$-neighborhood being all other datapoints that lie at a maximum distance of the specified $\epsilon$ to the starting point. If this neighborhood contains sufficiënt points, step 2 is started and if not, this point is marked as noise and step 1 is repeated.
2. If a point is found in a dense cluster, its $\epsilon$-neighborhood will also be taken as part of that cluster. This continues for eacht point added to the density cluster, until no new points are added.
3. The algorithm starts again at step 1, starting with an arbitray point that isn't marked as noise or as part of a density cluster. This is repeated until all datapoints are marked.

One of the benefits of this algorithm is the fact that the number of clusters does not have to be specified beforehand, as compared to KMeans where this is necessary. On the other hand, $\epsilon$ and the minimum amount of points in a cluster must be specified, which, in this case, can only be determined by trial and error, as there are no domain experts available for this dataset to provide these numbers.

Again, this algorithm is not completely deterministic. One of the reasons being the fact that a certain datapoint could be directly reached from different clusters.

[source](https://en.wikipedia.org/wiki/DBSCAN), [source](https://www.youtube.com/watch?v=C3r7tGRe2eI)

### Introduction

__LDA__ or __Latent Dirichlet Allocation__ is a statistical model that aims to find similarities in data proposed by [Blei et al](https://web.archive.org/web/20120207011313/http://jmlr.csail.mit.edu/papers/volume3/blei03a/blei03a.pdf). It is used to model how the topics of certain documents are related to eachother ([source](http://videolectures.net/mlss07_ghahramani_grafm/)).

[Blei et al](https://web.archive.org/web/20120207011313/http://jmlr.csail.mit.edu/papers/volume3/blei03a/blei03a.pdf) define LDA as "a three-level hierarchical Bayesian model, in which each item of a collection is modeled as a finite mxiture over an underlying set of topics. Each topic is in turn modeled as an infinite mixture over an underlying set of topic probabilities". Thus, these topic probabilities allow for a document to be represented explicitly in the context of document modeling and text classification.  

The model uses the __plate notation__, which is a way of representing variables that repeat in a graphical model. Here, variables are not repeated individually, but used in a _plate_ or _rectangle_ that groups variables into subgraphs that repeat together. 

<img src="images/plates.PNG" alt="plates" width="300"/>

The outer plate represents the actual documents, with _M_ being the number of documents. These include $\theta_d$, which stands for the document distribution of document _d_. 

The inner one represents the positions of repeated words in a document, $N$ (or $N_d$) being the number of words in the specified document $d$. Each of these positions is then associated with a word and assigned a chosen topic .$z_{d,n}$ is the topic of word $n$ in document $d$ and $w_{d,n}$ is the actual word that was used. The $K$ plate denotes the number of topics.

$\theta_d$ is the topic distribution for the document numbered as $d$ and $\beta_k$ represents the word distribution  for topic $k$.

Parameters $\eta$ and $\alpha$ are the hyperparameters for prior distributions of $\beta_k$ and $\theta_𝑑$, respectively. 

The arrows in the image indicate conditional dependancies.

In LDA, documents are represented as random combinations of hidden topics. Each of these topics is associated with their disctribution over all words. The LDA algorithm works as follow. 

With $D$ defining the corpus, that consists of $M$ documents that contain $N_i$ words:
1. Choose a $\theta_d$ that is proportional to $Dir(\alpha)$. $d$ being one of the $M$ documents and $Dir(\alpha)$ being the _Dirichlet distribution_ with a symmetric parameter $\alpha$.
2. Do the same with $\beta_k \sim Dir(\eta)$
3. For eacht of the word positions of word $w$ in document $d$
    1. Choose a topic $z_{d,n}\sim Multinomial(\theta_d)$, multinomial being the multinomial distribution, a generalization of the binomial distribution.
    2. Choose a word $w_{d,n}\sim Multinomial(\phi_{z_{d,n}})$

<img src="images/lda.PNG" alt="lda" width="500"/>

This way, LDA learns the importance of each topic and it can represent documents as a combination of their topics. This makes it easy to summarize documents based on the combinations of the topics that show their essence which can then be used to categorize the different document ([source](https://www.researchgate.net/publication/326505884_Latent_Dirichlet_Allocation_LDA_for_Topic_Modeling_of_the_CFPB_Consumer_Complaints)).

[Bastani et al](https://www.researchgate.net/publication/326505884_Latent_Dirichlet_Allocation_LDA_for_Topic_Modeling_of_the_CFPB_Consumer_Complaints) suggest a text mining approach based on LDA to analyse the consumer complaint narratives. [Aggarwal and Zhai](https://books.google.be/books?hl=en&lr=&id=vFHOx8wfSU0C&oi=fnd&pg=PR3&dq=Aggarwal,+C.+C.,+and+Zhai,+C.+(2012).+Mining+text+data&ots=oc9gYGjFSt&sig=ceG6vYV6rLFr8G_HkN5HREwHGGY#v=onepage&q=Aggarwal%2C%20C.%20C.%2C%20and%20Zhai%2C%20C.%20(2012).%20Mining%20text%20data&f=false) define text mining as the process of extracting non-trivial patterns or knowledge from unstructured text documents. These text documents are unstructured because they are inherently noisy. Short complaints about products tend more towards a collection of unstructured thoughts, rather than a coherent message. The difficulty lies in extract useful information that allows us to identify recurring patterns that could allow us to cluster these complaints and identify new complaints correctly.

Bastani et al suggest latent semantic analysis (LSA) as a potential approach in clustering and identifying narratives that have similar features. They state that LSA's shortcomings in effectively mining text lies in the fact that it assumes that a consumer complaint narrative belongs to a single topic. This is generally not the case, because a complaint may be assignable to multiple topics. Therefore, to combat this challenge, LDA is used to extract the latent topics from the consumer complaint narratives. According to Bastani et al, LDA assigns simulatanously assigns a probabilistic mixture of topics to the consumer complaint narratives. 

LDA could aid us in labeling complaints correctly and automatically, which would be otherwise be an immense work to do manually.

**Parameters** ([source](https://radimrehurek.com/gensim/models/ldamulticore.html)):
* alpha (α): Topic smoothing parameter; can be set to an 1D array of length equal to the number of expected topics that expresses our a-priori belief for the each topics’ probability
* eta (η): Word/term smoothing parameter; a scalar for a symmetric prior over topic/word probability

Most topic modeling analyses in the literature ([Blei et al, 2003](https://www.researchgate.net/publication/326505884_Latent_Dirichlet_Allocation_LDA_for_Topic_Modeling_of_the_CFPB_Consumer_Complaints); [Blei and Lafferty, 2009](https://citeseerx.ist.psu.edu/viewdoc/download?doi=10.1.1.463.1205&rep=rep1&type=pdf#page=96); [Kaplan and Vakili, 2015](https://onlinelibrary.wiley.com/doi/abs/10.1002/smj.2294); [Blei, 2012](https://dl.acm.org/doi/pdf/10.1145/2133806.2133826)) suggest a value of 0.1 for both of these hyperparameters. This results in semantically meaningful topics. However, these values can also be set to 'auto', meaning we would automatically learn these two parameters.

**Other parameters**:

* number of topics: The number of topics LDA has to attempt to identify
* iterations: Maximum number of iterations through the corpus when inferring the topic distribution of a corpus
* passes: Number of passes through the corpus during training
* chunksize: amount of documents processed at a time in the training algorithm

Increasing <code>chunksize</code> will speed up training, at least as long as the chunk of documents easily fit into memory. Chunksize can however influence the quality of the model, as discussed in Hoffman and co-authors ([source](https://papers.nips.cc/paper/3902-online-learning-for-latent-dirichlet-allocation.pdf)). <code>passes</code> controls how often we train the model on the entire corpus. Another word for passes might be “epochs”. <code>iterations</code> is somewhat technical, but essentially it controls how often we repeat a particular loop over each document.

**Optimal number of topics**

Finding the optimal number of topics can be done by building many LDA models with different values of number of topics (k) and pick the one that gives the highest coherence value.

Choosing a ‘k’ that marks the end of a rapid growth of topic coherence usually offers meaningful and interpretable topics. Picking an even higher value can sometimes provide more granular sub-topics.

Seeing the same keywords being repeated in multiple topics, probably indicates that the ‘k’ is too large.

The compute_coherence_values() (see below) trains multiple LDA models and provides the models and their corresponding coherence scores.

**Keywords and weightage**

We will also look at the **keywords** for each topic and the **weightage** (importance) of each keyword using lda_model.print_topics(). The output below show us what the top 10 keywords are that contribute to a topic. The weights reflect how important a keyword is to that topic.

**Model Perplexity** and **Topic Coherence**

Additionally, the **model perplexity** and **topic coherence** are printed. These provide a convenient measure to judge how good a given topic model is. Perplexity, however, appears to be a poor indicator of the quality of the topics ([source](https://datascienceplus.com/evaluation-of-topic-modeling-topic-coherence/)).

```python
from gensim import corpora, models
from gensim.models import CoherenceModel

# Training parameters
alpha = 0.1
eta = 0.1
#alpha = 'auto'
#eta = 'auto'
chunksize = 2000
iterations = 69
passes = 20

######################
# function that produces LDA models & computes coherence values
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=2):
    """
    Compute c_v coherence for various number of topics

    Parameters:
    ----------
    dictionary : Gensim dictionary
    corpus : Gensim corpus
    texts : List of input texts
    limit : Max num of topics

    Returns:
    -------
    model_list : List of LDA topic models
    coherence_values : Coherence values corresponding to the LDA model with respective number of topics
    """
    coherence_values = []
    model_list = []
    
    # Run model for different values for number of topics
    for num_topics in range(start, limit, step):
        print("Start with number of topics:", num_topics)
        # train
        lda_model = models.LdaMulticore(
                    corpus = corpus,
                    id2word = dictionary,
                    num_topics = num_topics, 
                    passes = passes,
                    iterations = iterations,
                    alpha = alpha, 
                    eta = eta,
                    random_state = 42
                   )
        # append to list of models
        model_list.append(lda_model)
        # save the trained model
        print("Saving the model...")
        lda_model.save("models/v2/lda_model_" + str(num_topics) + "topics")
        # Compute Perplexity
        print("Computing complexity...")
        print('Perplexity: ', lda_model.log_perplexity(corpus))  # a measure of how good the model is. lower the better.
        # Compute Coherence
        print("Computing coherence...")
        coherence_model_lda = CoherenceModel(model=lda_model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherence_model_lda.get_coherence())
        print('Coherence Score: ', coherence_model_lda.get_coherence())
        # Print the 10 most prominent topics for a given model
        for idx, topic in lda_model.print_topics(-1):
            print('\nTopic: {} \nWords: {}'.format(idx, topic))
        print("************************************************")
        
    return model_list, coherence_values
######################

# Run function (this will take some time)
model_list, coherence_values = compute_coherence_values(dictionary=id2word, 
                                                        corpus=corpus, 
                                                        texts=df['Consumer complaint narrative'].tolist(), 
                                                        start=8, 
                                                        step=2,
                                                        limit=40)
```

**Dominant topic and its percentage contribution to each document (complaint)**.

The table below shows the dominant topic and its percentage contribution for a given complaint (text), as well as the keywords that define this text according to the model. The model illustrated here is LDA_v2 with 14 topics.

<img src='images/dominant_topic_and_perc_contri_each_doc.png' />

```python
def format_topics_sentences(ldamodel, corpus, texts):
    # Init output
    complaints_topics_df = pd.DataFrame()

    # Get main topic in each document
    for i, row_list in enumerate(ldamodel[corpus]):
        row = row_list[0] if ldamodel.per_word_topics else row_list            
        # print(row)
        row = sorted(row, key=lambda x: (x[1]), reverse=True)
        # Get the Dominant topic, Perc Contribution and Keywords for each document
        for j, (topic_num, prop_topic) in enumerate(row):
            if j == 0:  # => dominant topic
                wp = ldamodel.show_topic(topic_num)
                topic_keywords = ", ".join([word for word, prop in wp])
                complaints_topics_df = complaints_topics_df.append(pd.Series([int(topic_num), round(prop_topic,4), topic_keywords]), ignore_index=True)
            else:
                break
    complaints_topics_df.columns = ['Dominant_Topic', 'Perc_Contribution', 'Topic_Keywords']

    # Add original text to the end of the output
    contents = pd.Series(texts)
    complaints_topics_df = pd.concat([complaints_topics_df, contents], axis=1)
    
    return(complaints_topics_df)


df_topic_sents_keywords = format_topics_sentences(ldamodel=lda_model14, corpus=corpus, texts=df['Consumer complaint narrative'].reset_index())

# Format
df_dominant_topic = df_topic_sents_keywords.reset_index()
df_dominant_topic.columns = ['Document_No', 'Dominant_Topic', 'Topic_Perc_Contrib', 'Keywords', 'Text']
df_dominant_topic.head(10)
```

#### Conclusion

**v2 runs** did not offer that much improvement over **v1 runs** in terms of coherence scores & model perplexity, despite the additional stopwords. However, it could be stated that these v2_runs are still better, because the topical keywords appeared to be more relevant and convey more information about the topic. This then allows for an expert to more clearly label said topics.

**v3 runs** had a higher coherence score, indicating that these could perhaps be better than the v2 runs. The topics were indeed relevant and clear and distinct topics (> 20) could still be found, indicating the overlap may have. For example, the v3 model with 22 topics shows significant overlap on the pyLDAvis, however, this could be because certain keywords such as 'credit' appear in both a topic such as *credit card* and a topic such as *credit for loans*. Or a keyword such as 'loan' appearing in both a topic such as 'student loans' and 'mortgage/house loans'. Including bigrams could potentially help fix these overlaps. 

We evaluated the LDA models using coherence score. The overall coherence score of a topic is the average of the distances between words. For c_v, which is what was used, this value is 0 < x < 1. This score increased as more topics were added, but these topics may have repeated keywords in the topic. Possible low coherence fixes could involves adjusting the parameters such as alpha, cleaning the data even more, or using a better LDA algorithm such as Mallet's implementation. Realistically, more work needs to be done to boost this score, because the best score for these runs was around 0.52. Alternatively, we could try the U mass measure instead of the c_v measure.

We also looked at perplexity as an indicator for our model, but this is a poor indicator of the quality of the topics. Thus, we visualised the topic model using wordclouds and PyLDAvis. 

We labeled the v2 model with 14 topics as in illustration of what could be done now. It would now be up to a field expert to help clearly label these topics and determine whether or not more runs are needed, or whether fewer or more topics could still be added.
